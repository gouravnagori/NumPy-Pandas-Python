```python
df.dropna() # row drop
```

```python
# if you want to drop column 
df.dropna(axis = 1) #column drop
```

```python
df.fillna("Empty")
df.fillna(96) #we can fill the null value with any value
# but it is not good practice to fill null value with any value so we have to be specific
```

```python
# so if we want to fill NaN value of age with a meaningful value so
# can we replace it with mean value????
# Yes!!
age_mean = df["age"].mean()
df["age"].fillna(age_mean)
```

```python
cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data
```

```python
# Melt and pivot
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Country": ["USA" , "IND" , "AUS" , "ENG" , "CHINA"] ,
    "Avg_Age": [60,55,73,74,80],
    "Year"   : [2022,2022,2021,2025,2024] ,
    "Income" : [90000,15000,80000,97332,99000],
    "Tax"    : [12,18,12,10,8]
})
```

```python
df
```

```python
melted_data = df.melt( id_vars = ["Country" , "Year"],
     value_vars = ["Income","Tax","Avg_Age"],
     var_name = "Total_Salary_with_Age",
     value_name = "value")
melted_data 
```

```python
original_data = melted_data.pivot(
    index = ["Country"] ,
    columns = "Total_Salary_with_Age",
    values =   "value"      
)
```

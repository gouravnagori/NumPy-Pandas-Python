```python
import numpy as np
```

```python
# aggregation
arr = np.array([1,4,6,0,2,3,1])
print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr)) #return index no. of the minimum value
print(np.argmax(arr)) #return index no. of the maximum value
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))


```

```python
# Power function
print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python
# Log and exp function
arr = np.array([1,2,-310,-34,10,100,2,3])
print(np.log(arr))  #by default base e
print(np.log10(arr))
print(np.log2(arr))
print(np.exp(arr))
```

```python
# Rounding
print(np.round(3.14))
print(np.round(3.4))
print(np.round(3.879))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.round(3.14))
print(np.trunc(3.14))


# unique function
print(np.unique(arr))
print(np.sort(arr))
print(np.abs(arr))
```

```python
import pandas as pd
```

```python
# DataFrame

# here we create DatFrame with the help of dict
info = {
    "Name" : ["Ram" , "Shyam" , "Bharat" , "Laxman"],
    "Age"  : [1000,1000,1000,1000] ,
    "Power": ["Calmness" , "Kindness" , "Respectful" , "Archery"]
}
df = pd.DataFrame(info)
print(df)
print(type(df))
print(df.index)
print(df.columns)
```

```python
# we can also create DataFrame with the help of list of list
df2 = pd.DataFrame([["rahul" ,19] , ["raj" , 19]] , columns = ["Name" , "Age"])
print(df2)
```

```python
# we can also create Datframe with the helpof NumPy array
import numpy as np
np_arr = np.array([["Harsh" ,23 , "Engg"] , ["Hitesh" ,30 , "Doct"]])
df3 = pd.DataFrame(np_arr , columns = ["Name" , "Age" , "Occup"])
print(df3)
```

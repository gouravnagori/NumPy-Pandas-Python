```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4,5,6,7,8])
print(s)
print(type(s))
```

```python
# Index based access
print(s[0])
print(s[3])
print(s[2])
```

```python
# Label based access
# We can understand lable like index but we can create our own label
# we can define label name in index attribute
s = pd.Series([23,43,21,2,5] , index = ["Rahul" , "Neha" , "Aryan" , "Joe" , "Adam"])
print(s)
print(s["Neha"])
print(s["Joe"]) # access via label
print(s[0]) #try to access the value via index and we got worning not an error
print(s.index)
```

```python
ser1 = pd.Series([1,2,3,4,5])
ser2 = pd.Series([10,20,30,40,50])
print(ser2-ser1)
```

```python
ser1 = pd.Series([1,2,3,4,5])
ser2 = pd.Series([10,20,30,40,50])
print(ser2-ser1)

ser1[0] = 69
print(ser1)
changed_ser1 = ser1.drop(2)
print(changed_ser1)
```

```python
# Common Data Types
```

```python
import numpy as np
```

```python
arr = np.array([2,4,2,4,7])
print(arr,arr.dtype)

arr = np.array([2,4,2.2,4,7])
print(arr,arr.dtype)

arr = np.array([2+4j,2.2,4,7])
print(arr,arr.dtype)

arr = np.array([2,4,'AI/ML',2.2,4,7])
print(arr,arr.dtype)

arr = np.array([True,False])
print(arr,arr.dtype)

arr = np.array([2,4,None,'AI/ML',2.2,4,7])
print(arr,arr.dtype)
```

```python
arr1 = np.array([2+5j])
arr2 = np.array([1+3J])
print(arr1 + arr2)
sum = arr1 + arr2
print(sum.dtype)
```

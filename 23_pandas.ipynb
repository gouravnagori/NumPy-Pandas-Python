## concatenation

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "id" : [1,2,3,4] , 
    "name": ["Rahul" , "Diplip" , "Jayant" ,"Dev" ]
})

df2 = pd.DataFrame({
    "id" : [5,6,7] , 
    "name": ["Neha" , "Aditi" , "Alex" ]
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1 , df2])  #row wise
```

```python
pd.concat([df2 , df1] , ignore_index=True)  #row wise
```

```python
pd.concat([df1 , df2] , axis = 1)  #col wise
```

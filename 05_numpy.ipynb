```python
# Multi-dimension Array
```

```python
import numpy as np
```

```python
arr2D  = np.array([[1,2,3],[3,3,1],[9,4,1]]) 
print(arr2D)
```

```python
print("Sum of all element of matrix",np.sum(arr2D))
```

```python
print("Sum of columns",np.sum(arr2D , axis = 0))
```

```python
print("Sum of rows",np.sum(arr2D , axis = 1))
```

```python
# Slicing concept
# here also we can get particular part of an array
# like if i want to print the only 3 and 4 value from my array name arr2D so that i can access them like that
print(arr2D[1: , 1:2])
```

```python
# 3-D Array
arr3D = np.array([[[1,2] , [2,0] , [7,4]] ,[[5,4] , [3,5] , [9,0]]])
print(arr3D , arr3D.shape)
print(np.sum(arr3D))

# Indexing
print(arr3D[0,2,1])

# Slicing
print("first row from both the layer  ",arr3D[:,0,:])
print("first column from both the layer  ",arr3D[:,:,0])

# Data manipulation
arr3D[:,:,0] = 99
print(arr3D)
```

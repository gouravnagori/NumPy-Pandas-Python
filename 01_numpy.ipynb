```python
# execution performance

size = 10_000_000

#python list
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"Time taken by python list is {end-start} seconds")

#NumPy array
np_arr = np.array(py_list)
start = time.time()
sq_arr = np_arr ** 2
end = time.time()
print(f"Time taken by NumPy array is {end-start} seconds")
```

```python
# Memory occupation

import sys
import numpy as np

py_list = [1,3,5,23,4,544,56,3,54,4,23,4,5,3,4,3,3,5,6,67,4,23,4,67,4,23,5,6]
np_arr = np.array(py_list)

list_memory = sys.getsizeof(py_list)
int_memory = sum(sys.getsizeof(x) for x in py_list)

print("List object size or container size:", list_memory,"bytes")
print("All integers size:", int_memory,"bytes")
print("Total Python list size:", list_memory + int_memory,"byts")
print("NumPy array size:", np_arr.nbytes,"bytes")
```

Conclusion: Python List vs NumPy Array (Memory Usage)

Python List
Stores references (pointers) to individual Python objects.
Each element is a separate Python object with additional memory overhead.
sys.getsizeof(py_list) returns the size of the list container only, not the total memory used by all elements.

NumPy Array
Stores elements contiguously in memory with the same data type.
Uses much less memory because there is no per-element Python object overhead.
np_arr.nbytes returns the actual memory occupied by the array's data.

Important Difference
len(py_list) → Returns the number of elements in the list.
sys.getsizeof(py_list) → Returns the memory of the list object (container) only.
np_arr.nbytes → Returns the memory used by all elements in the NumPy array.

Final Conclusion
NumPy arrays are generally more memory-efficient and faster than Python lists for numerical data because they store values directly in a continuous block of memory. Python lists consume more memory due to the overhead of storing each element as a separate Python object.

Pro Tip: Never compare sys.getsizeof(py_list) directly with np_arr.nbytes. getsizeof() measures only the list container, whereas nbytes measures the entire data stored in the NumPy array. For a fair comparison, include the memory of the list's individual elements as well.

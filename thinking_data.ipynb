```python
import json
#load the data
def load_data(file):
    with open(file , "r") as f:
        data = json.load(f)
    return data
    
```

```python
data = load_data("store_data.json")
print(data)
```

```python
#current rating
for user in data:
    print(user["rating"])
```

```python
# clean and structure the data
def clean_data(data):
    unorder_to_order = {
        'one':1,
        'two':2,
        'three':3,
        'four':4,
        'five':5
    }
    cleaned_data =[]
    unique_user = set()
    for user in data:
        #clean rating data
        order_rating = str(user["rating"].strip().lower())
        if order_rating in unorder_to_order:
            order_rating = unorder_to_order[order_rating]
        print(order_rating)   
        # overwrite the data of rating
        user["rating"] = order_rating

        # handle missing values
        raw_age = user.get("age")
        if raw_age == None:
            user["age"] = None

        # remove duplicacy
        if user["name"].strip() in unique_user:
            continue
            
        unique_user.add(user["name"])    
        cleaned_data.append(user)
    print(cleaned_data)      
```

```python
clean_data(data)
```

```python
# meaningful insights from the clean data

def get_insights(data):
    # Avg. rating
    sum = 0
    for user in data:
        int_rating = float(user["rating"])
        sum = int_rating + sum
    print(f"The avg rating is {sum/len(data)}")    


    #percentage of user with poor ratings (3>)
    poor_ratings = 0
    for user in data:
          if float(user["rating"]) < 3:
              poor_ratings += 1
    print(f"{(poor_ratings/len(data)) * 100}% user give poor rating")
    
```

```python
get_insights(data)
```

```python
#Build recommendation Feature
def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]

        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"

        recommendations.append(curr_recomm)
    print(recommendations)

```

```python
get_recommendations(data)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df2 = df.copy()
```

```python
df2.columns = ["ID", "NAME" , "AGE" , "COUNTRY" , "GENDER" , "INCOME"]
```

```python
df2
```

```python
# INCOME ---> Salary
df2.rename(columns = {"INCOME" : "Salary"})
```

```python
# if want to rename the lable name
df2.rename(index = {0 : "Zero" , 1 : "first"})
```

```python
df2 = df2.fillna(50)
# sort_inc = df2.sort_values("INCOME" , ascending =False)
sorted_data = df2.sort_values(["INCOME" , "AGE"])
sorted_data
```

```python
sorted_data.sort_index()
```

```python
# reset
sorted_data.reset_index()

# and we can also drop our old index by doing
sorted_data.reset_index(drop = True)
```

```python
# ranking our data
sorted_data["Ranking"] = sorted_data["INCOME"].rank(ascending = False)
sorted_data
# and you don't want that kind of values like 6.5 and 6.5 then you can change it by doing this
sorted_data["Ranking"] = sorted_data["INCOME"].rank(ascending = False , method = "dense")
sorted_data
```

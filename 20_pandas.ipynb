```python
import pandas as pd
```

```python
df = pd.read_csv("sorted_data.csv")
df
```

```python
df.groupby("gender")["age"].mean()
```

```python
df.groupby("gender")["income"].max()
```

```python
df.groupby("country")["age"].min()
```

```python
df.groupby("country")["income"].agg(["min","mean","max"])
```

```python
# .aggregate() is same as .agg()
df.groupby("country")["income"].aggregate(["min","mean","max"])
```

```python
df.groupby("country")["income"].aggregate(min_income = "min",avg_income = "mean",max_income = "max")
```

```python
df.groupby("country").agg({
    "income" : "mean",
    "age" : "min"
})
```

```python
df.groupby("gender").agg({
    "income" : "mean",
    "age" : "mean"
})
```

```python
df.groupby("country").agg(
    max_salary=("income", "max"),
    avg_age=("age", "mean")
)
```
